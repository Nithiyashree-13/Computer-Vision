***Task 1: Simple python code to read an Image and Convert it into a NumPy array. Display the image of only red channel, blue channel, green channel separately and then converting to grayscale***

```python
image = cv2.imread('/content/Bird.jpg')
display(Image('/content/Bird.jpg'))
```

```python
numpy_array = np.array(image)
print (numpy_array)
```

```python
print(image.shape)

# extract red channel
red_channel = image[:,:,2]

# create empty image with same shape as that of the original image
red_img = np.zeros(image.shape)

#assign the red channel of the original image to empty image
red_img[:,:,2] = red_channel

#save image
cv2.imwrite('C:\\Users\\Nithi\\Desktop\\Red_Channel.png',red_img) 
display(Image('C:\\Users\\Nithi\\Desktop\\Red_Channel.png'))
```

```python
# extract green channel
green_channel = image[:,:,1]

# create empty image with same shape as that of the original image
green_img = np.zeros(image.shape)

#assign the green channel of the original image to empty image
green_img[:,:,1] = green_channel

#save image
cv2.imwrite('C:\\Users\\Nithi\\Desktop\\Green_Channel.png',green_img) 
display(Image('C:\\Users\\Nithi\\Desktop\\Green_Channel.png'))
```

```python
# extract blue channel
blue_channel = image[:,:,0]

# create empty image with same shape as that of the original image
blue_img = np.zeros(image.shape)

#assign the red channel of the original image to empty image
blue_img[:,:,0] = blue_channel

#save image
cv2.imwrite('C:\\Users\\Nithi\\Desktop\\Blue_Channel.png',blue_img) 
display(Image('C:\\Users\\Nithi\\Desktop\\Blue_Channel.png'))
```

```python
# converting a color image to a grayscale image
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImage)
```
